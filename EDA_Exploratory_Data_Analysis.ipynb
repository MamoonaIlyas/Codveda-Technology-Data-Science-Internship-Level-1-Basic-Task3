```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for better aesthetics
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Default figure size for plots

# Load the dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

print("--- Original Dataset Info ---")
df.info()
print("\nFirst 5 rows of the dataset:")
print(df.head())
print("\n" + "="*80 + "\n")

# --- 1. Compute Summary Statistics ---
print("--- 1. Summary Statistics ---")

# A. Basic descriptive statistics for all numerical columns
# This provides count, mean, std, min, 25%, 50% (median), 75%, max
print("Descriptive Statistics for Numerical Features:")
print(df.describe())
print("\n")

# B. Specific statistics for individual columns (example for 'CRIM' and 'MEDV')
print("Mean of 'CRIM' (per capita crime rate):", df['crim'].mean())
print("Median of 'MEDV' (Median value of owner-occupied homes in $1000s):", df['medv'].median())
print("Variance of 'LSTAT' (% lower status of the population):", df['lstat'].var())
print("Standard Deviation of 'RM' (average number of rooms per dwelling):", df['rm'].std())
print("\n" + "="*80 + "\n")


# --- 2. Visualize the Data ---

print("--- 2. Data Visualizations ---")

# A. Histograms for Numerical Features
# Histograms show the distribution of individual numerical variables.
print("Generating Histograms...")
df.hist(bins=30, figsize=(15, 10), layout=(4, 4))
plt.suptitle('Histograms of Boston Housing Features', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
print("Histograms displayed. Look for skewness, multiple peaks, or unusual distributions.\n")

# B. Scatter Plots for Key Relationships
# Scatter plots help visualize relationships between two numerical variables.
print("Generating Scatter Plots for key relationships...")

# Relationship between 'RM' (rooms) and 'MEDV' (median house value)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='rm', y='medv', data=df, alpha=0.7)
plt.title('Rooms (RM) vs. Median House Value (MEDV)')
plt.xlabel('Average number of rooms per dwelling (RM)')
plt.ylabel('Median value of owner-occupied homes in $1000s (MEDV)')
plt.show()

# Relationship between 'LSTAT' (% lower status) and 'MEDV'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='lstat', y='medv', data=df, alpha=0.7)
plt.title('Lower Status Population (LSTAT) vs. Median House Value (MEDV)')
plt.xlabel('% lower status of the population (LSTAT)')
plt.ylabel('Median value of owner-occupied homes in $1000s (MEDV)')
plt.show()

# Relationship between 'CRIM' (crime rate) and 'MEDV'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='crim', y='medv', data=df, alpha=0.7)
plt.title('Crime Rate (CRIM) vs. Median House Value (MEDV)')
plt.xlabel('Per capita crime rate by town (CRIM)')
plt.ylabel('Median value of owner-occupied homes in $1000s (MEDV)')
plt.show()
print("Scatter plots displayed. Look for linear, non-linear, or no clear patterns.\n")

# C. Box Plots for Distribution and Outliers
# Box plots are great for visualizing distribution and identifying outliers.
print("Generating Box Plots for selected features...")

# Box plot for 'MEDV'
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['medv'])
plt.title('Box Plot of Median House Value (MEDV)')
plt.ylabel('Median value of owner-occupied homes in $1000s (MEDV)')
plt.show()

# Box plot for 'RM'
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['rm'])
plt.title('Box Plot of Average Number of Rooms (RM)')
plt.ylabel('Average number of rooms per dwelling (RM)')
plt.show()

# Box plot for 'CRIM' (often has many outliers)
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['crim'])
plt.title('Box Plot of Per Capita Crime Rate (CRIM)')
plt.ylabel('Per capita crime rate by town (CRIM)')
plt.show()
print("Box plots displayed. Look for central tendency, spread, and individual outlier points (dots).\n")
print("\n" + "="*80 + "\n")


# --- 3. Identify Correlations Between Numerical Features ---

print("--- 3. Correlation Matrix (Heatmap) ---")

# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Boston Housing Features', fontsize=16)
plt.show()
print("Correlation heatmap displayed. Values closer to 1 or -1 indicate stronger linear relationships.\n")
print("\n" + "="*80 + "\n")


# --- 4. Generate a Report Summarizing Insights from EDA ---

print("--- 4. EDA Insights Report ---")
print("Based on the Exploratory Data Analysis of the Boston Housing dataset, here are key insights:\n")

print("1.  **Dataset Overview:**")
print("    - The dataset contains 506 entries and 14 features, including the target variable 'MEDV' (Median value of owner-occupied homes).")
print("    - All columns are numerical and initially have no missing values, indicating a clean dataset for analysis.")
print("    - Features like 'CRIM', 'ZN', 'CHAS', 'RAD', 'B' are integer types, while others are float, suggesting varying scales and types of data.")

print("\n2.  **Feature Distributions (from Histograms and Box Plots):**")
print("    - **'MEDV' (Target Variable):** Appears to be somewhat normally distributed but shows a slight left skew, with a noticeable peak around $20,000-$25,000. There are also some values capped at $50,000, which might indicate a data collection limitation or censoring.")
print("    - **'CRIM' (Crime Rate):** Highly right-skewed, with most towns having very low crime rates and a few towns having significantly high rates (many outliers). This suggests a non-linear relationship with other variables and might benefit from transformation.")
print("    - **'ZN' (Residential Land Zoned):** Also highly right-skewed, indicating most properties are not in areas zoned for large lots.")
print("    - **'CHAS' (Charles River Dummy Variable):** A binary variable (0 or 1), showing that only a small proportion of properties are adjacent to the Charles River.")
print("    - **'RM' (Average Rooms):** Shows a relatively normal distribution, but with some outliers on both the lower and higher ends, representing unusually small or large homes.")
print("    - **'LSTAT' (% Lower Status):** Right-skewed, indicating more areas with lower percentages of lower-status population, but also a tail towards higher percentages.")
print("    - **'DIS' (Distances to Employment Centers):** Right-skewed, with most properties closer to employment centers.")

print("\n3.  **Relationships Between Features (from Scatter Plots and Correlation Matrix):**")
print("    - **'RM' vs. 'MEDV':** Shows a strong positive linear correlation (~0.70). As the average number of rooms increases, the median house value tends to increase. This is a very important predictor.")
print("    - **'LSTAT' vs. 'MEDV':** Exhibits a strong negative linear correlation (~-0.74). As the percentage of lower-status population increases, the median house value tends to decrease. This is also a very strong predictor.")
print("    - **'CRIM' vs. 'MEDV':** Shows a negative correlation (~-0.39), suggesting that higher crime rates are generally associated with lower house values, though the relationship is not as strong or linear as 'RM' or 'LSTAT'. The scatter plot confirms a wide spread, especially at lower crime rates.")
print("    - **Multicollinearity:**")
print("        - 'NOX' (nitric oxides concentration) is positively correlated with 'INDUS' (non-retail business acres) (~0.76) and 'AGE' (proportion of owner-occupied units built prior to 1940) (~0.74). This makes sense as older industrial areas might have higher pollution.")
print("        - 'TAX' (full-value property-tax rate) and 'RAD' (index of accessibility to radial highways) are highly positively correlated (~0.91). This suggests that areas with higher accessibility to highways also tend to have higher property taxes, or vice-versa.")
print("        - 'DIS' (weighted distance to employment centers) is negatively correlated with 'INDUS' (~-0.71) and 'NOX' (~-0.77). This is logical: areas further from employment centers might be less industrialized and have lower pollution.")

print("\n5.  **Potential Issues/Next Steps:**")
print("    - **Outliers:** Several features, particularly 'CRIM', 'ZN', and 'B', show significant outliers. 'MEDV' also has a cap at $50,000. These outliers might need handling (removal, transformation, or capping) to prevent them from disproportionately influencing models.")
print("    - **Skewness:** Highly skewed features like 'CRIM', 'ZN', 'DIS', and 'LSTAT' might benefit from log transformations or other power transformations to make their distributions more symmetrical, which can improve the performance of some linear models.")
print("    - **Multicollinearity:** The high correlation between 'TAX' and 'RAD' (and others like 'NOX'/'INDUS'/'AGE'/'DIS') suggests potential multicollinearity. This should be considered when choosing models (e.g., linear regression is sensitive to it; tree-based models less so) or by using techniques like PCA.")
print("    - **Feature Engineering:** The 'CHAS' variable (binary) might be directly useful. Creating interaction terms (e.g., 'RM' * 'LSTAT') could capture more complex relationships.")
print("\nEDA provides a solid foundation for subsequent data preprocessing and model building. The strong correlations of 'RM' and 'LSTAT' with 'MEDV' suggest they will be important features for prediction.")
```
